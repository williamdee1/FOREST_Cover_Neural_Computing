```python
import numpy as np
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('ggplot')
```

# Load Data:

```python
for_cover = pd.read_csv('Data/covtype.csv')
```

```python
for_cover.head(10)
```

```python
cover_dict = {1 : "Spruce/\n Fir", 2 : "Lodgepole \nPine", 3 : "Ponderosa \nPine", 4 : "Cottonwood/\n Willow", 
              5 : "Aspen", 6 : "Douglas-fir", 7 : "Krummholz"}
```

```python
for_cover['cover_name'] = for_cover['Cover_Type'].map(cover_dict)
```

# Visualizing Target Class Imbalance:

```python
cov_type_count = for_cover.Cover_Type.value_counts().reset_index().rename(columns={'index': 'cover_num'}).sort_values('cover_num')
cov_type_count['cov_name'] = cov_type_count['cover_num'].map(cover_dict)
cov_type_count
```

```python
plt.figure(figsize=(12,6))
plt.rcParams['axes.facecolor'] = 'white'
colors = ['darkgreen', 'darkkhaki', 'lightblue', 'olive', 'mediumseagreen', 'salmon', 'rosybrown']
bars = plt.bar(cov_type_count.cov_name, cov_type_count.Cover_Type)

# Setting bar colour individually:
for i in range (len(cov_type_count.cov_name)):
    bars[i].set_color(colors[i])

# Annotating count number above bars:
for j, val in enumerate(cov_type_count.Cover_Type):
    plt.text(j, val, f'{val:,}', horizontalalignment='center', verticalalignment='bottom', 
             fontdict={'fontweight':500, 'size':12}, c= 'black')

plt.title('Dataset Instances of Different Forest Cover Types', fontweight = 'bold')
plt.yticks([])
# removing tick marks - https://stackoverflow.com/questions/29988241/python-hide-ticks-but-show-tick-labels
plt.tick_params(axis=u'both', which=u'both',length=0) 
plt.xticks(fontsize=13)
plt.grid()
#plt.savefig('Figures/Cover_Type_Dist.png', bbox_inches='tight')
plt.show()
```

# Summary Statistics of Non-Binary Predictor Variables:

```python
non_binary_vars = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology',
       'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',
       'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
       'Horizontal_Distance_To_Fire_Points']
```

# Box Plots of Distribution of Non-Binary Variables:

```python
customPalette = sns.set_palette(sns.color_palette(colors))
```

```python
fig, axes = plt.subplots(5, 2, figsize=(15,20))
axes = axes.flatten()
plt.rcParams['axes.facecolor'] = 'white'


for i, col in enumerate(for_cover[non_binary_vars]):
    sns.boxplot(x=for_cover['cover_name'], y=col, data=for_cover, palette=customPalette, 
                ax = axes[i], order = cov_type_count.cov_name)
    axes[i].set(xlabel=None, ylabel=None)
    axes[i].tick_params(axis=u'both', which=u'both',length=0)
    axes[i].set_title('%s' %(col), fontweight='bold', fontsize=12)
plt.tight_layout()
plt.grid()
plt.show()
```

- We can see variables such as Elevation are likely very strong predictors as the distribution is clearly separated amongst the target classes.

## Soil Types:

```python
soil_vars = ['Soil_Type1', 'Soil_Type2',
 'Soil_Type3', 'Soil_Type4', 'Soil_Type5', 'Soil_Type6', 'Soil_Type7', 'Soil_Type8', 'Soil_Type9', 'Soil_Type10', 'Soil_Type11',
 'Soil_Type12', 'Soil_Type13', 'Soil_Type14', 'Soil_Type15', 'Soil_Type16', 'Soil_Type17', 'Soil_Type18',
             'Soil_Type19', 'Soil_Type20', 'Soil_Type21', 'Soil_Type22', 'Soil_Type23', 'Soil_Type24', 'Soil_Type25', 
                'Soil_Type26', 'Soil_Type27', 'Soil_Type28', 'Soil_Type29', 'Soil_Type30', 'Soil_Type31', 'Soil_Type32', 
                'Soil_Type33', 'Soil_Type34', 'Soil_Type35', 'Soil_Type36', 'Soil_Type37', 'Soil_Type38', 'Soil_Type39',
                'Soil_Type40']
```

```python
soil_count_df = for_cover.groupby('Cover_Type')[soil_vars].sum().T.reset_index(drop=True)
```

```python
fig = plt.figure(figsize=(8,4))
plt.rcParams['axes.facecolor'] = 'white'
soil_types = list(range(0, len(soil_count_df)))

plt.bar(soil_types, soil_count_df[1], color=colors[0], edgecolor='white', width=1)

for i in range(2, 8):
    plt.bar(soil_types, soil_count_df[i], bottom = soil_count_df[i-1], color=colors[i-1], width=1)
    

plt.grid()
plt.show()
```

- Some soil types have very little data in them, however we will experiment with dropout in the MLP model to see if it has an overall positive impact. 
- Both chosen models deal well with high dimensionality in the data and so should be okay to deal with the number of variables included in this dataset.
